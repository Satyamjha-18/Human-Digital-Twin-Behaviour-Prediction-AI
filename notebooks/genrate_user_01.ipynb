```python
import random
import pandas as pd

# -------------------------
# 1. Basic Lists
# -------------------------

first_names = [
    "Aarav", "Vivaan", "Aditya", "Arjun", "Rohan", "Rahul",
    "Aman", "Karan", "Sahil", "Vikram", "Neha", "Pooja",
    "Sneha", "Priya", "Anjali", "Kavya", "Riya", "Simran"
]

last_names = [
    "Sharma", "Verma", "Singh", "Gupta", "Patel",
    "Yadav", "Mehta", "Kumar", "Jain", "Shah"
]

occupations = [
    "Student",
    "Software Engineer",
    "Doctor",
    "Teacher",
    "Data Analyst",
    "Business Owner",
    "Accountant",
    "Lawyer",
    "Designer",
    "Sales Executive"
]

cities = [
    "Delhi", "Mumbai", "Bhopal", "Pune", "Bengaluru",
    "Hyderabad", "Jaipur", "Indore", "Lucknow", "Chennai"
]

# -------------------------
# 2. Helper Functions
# -------------------------

def generate_age(occupation):
    if occupation == "Student":
        return random.randint(18, 25)
    elif occupation == "Software Engineer":
        return random.randint(22, 40)
    elif occupation == "Doctor":
        return random.randint(27, 60)
    elif occupation == "Teacher":
        return random.randint(25, 60)
    elif occupation == "Data Analyst":
        return random.randint(22, 45)
    elif occupation == "Business Owner":
        return random.randint(25, 65)
    else:
        return random.randint(22, 55)

def generate_height():
    return round(random.uniform(150, 190), 1)

def generate_weight(height):
    # simple BMI logic
    bmi = random.uniform(18.5, 28)
    weight = bmi * ((height/100) ** 2)
    return round(weight, 1)

# -------------------------
# 3. Generate Users
# -------------------------

data = []

for i in range(1, 1001):

    first = random.choice(first_names)
    last = random.choice(last_names)
    name = first + " " + last

    occupation = random.choice(occupations)
    age = generate_age(occupation)

    height = generate_height()
    weight = generate_weight(height)

    city = random.choice(cities)

    user = {
        "user_id": i,
        "name": name,
        "age": age,
        "gender": random.choice(["Male", "Female"]),
        "height_cm": height,
        "weight_kg": weight,
        "occupation": occupation,
        "city": city
    }

    data.append(user)

# -------------------------
# 4. Save to CSV
# -------------------------

df = pd.DataFrame(data)
df.to_csv("users.csv", index=False)

print("✅ 1000 Users generated successfully!")
```
