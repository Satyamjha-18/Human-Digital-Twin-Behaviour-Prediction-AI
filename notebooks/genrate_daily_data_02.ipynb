```python
records = []

for _, user in users.iterrows():

    base_sleep = np.random.uniform(6, 8)
    base_work = np.random.uniform(6, 9)

    for date in dates:

        # Weekend effect
        weekend = date.weekday() >= 5

        sleep = np.clip(base_sleep + np.random.normal(0, 0.5), 3, 9)

        if weekend:
            work = np.clip(np.random.normal(4, 1), 0, 8)
        else:
            work = np.clip(base_work + np.random.normal(0, 1), 4, 12)

        screen = np.clip(np.random.normal(5.5, 2), 1, 12)

        steps = int(np.clip(np.random.normal(7000, 2500), 1000, 20000))

        exercise = int(np.clip(np.random.normal(35, 20), 0, 120))

        water = round(np.clip(np.random.normal(2.5, 0.5), 1, 5), 1)

        # AI Behavior Rules
        stress = (
            screen * 0.35 +
            work * 0.30 -
            sleep * 0.40 -
            exercise * 0.02
        )

        stress = np.clip(stress, 1, 10)

        mood = (
            sleep * 0.50 +
            exercise * 0.04 -
            stress * 0.35
        )

        if weekend:
            mood += 0.5

        mood = np.clip(mood, 1, 10)

        productivity = (
            work * 0.40 +
            mood * 0.45 -
            stress * 0.20
        )

        productivity = np.clip(productivity, 1, 10)

        records.append([
            user["user_id"],
            date.strftime("%Y-%m-%d"),
            round(sleep, 1),
            round(work, 1),
            round(screen, 1),
            steps,
            exercise,
            water,
            round(stress, 1),
            round(mood, 1),
            round(productivity, 1)
        ])
```

```python
columns = [
    "user_id",
    "date",
    "sleep_hours",
    "work_hours",
    "screen_time",
    "steps",
    "exercise_minutes",
    "water_intake",
    "stress_level",
    "mood_score",
    "productivity_score"
]

daily_behavior = pd.DataFrame(records, columns=columns)

daily_behavior.head()
```
